# importing seaborn and matplot libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading iris dataset

```python
iris=sns.load_dataset('iris')
```

```python
print(iris)
```

# General Statistics Plot

```python
sns.pairplot(iris,hue='species',height=2.3)
plt.show()
```

# Pie Plot for Species Frequency:

```python
spe_counts = iris['species'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(spe_counts, labels=spe_counts.index, autopct='%1.1f%%', startangle=0)
plt.title('Species Frequency in Iris Dataset')
plt.show()
```

# Relationship Between Sepal Length and Width:

# sepal length vs sepal width

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

# petal length vs petal width

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

# Distribution of Sepal and Petal Features:

```python
sns.pairplot(iris, hue='species', height=2.5)
plt.show()
```

# Jointplot of Sepal Length vs Sepal Width:

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter',color='pink')
plt.show()
```

# KDE Plot for Setosa Species (Sepal Length vs Sepal Width):

```python
setosa = iris[iris['species'] == 'setosa']
sns.kdeplot(x='sepal_length', y='sepal_width', data=setosa, fill=True,color='red')
plt.title('KDE Plot of Sepal Length vs Sepal Width (Setosa)')
plt.show()
```

# KDE Plot for Setosa Species (Petal Length vs Petal Width):

```python
sns.kdeplot(x='petal_length', y='petal_width', data=setosa, fill=True,color='green')
plt.title('KDE Plot of Petal Length vs Petal Width (Setosa)')
plt.show()
```
